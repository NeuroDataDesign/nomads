# 1. Description
Non-negative matrix factorization (NMF) is a method of decomposing a matrix into 2 smaller matrices, both of which are strictly positive, that are multiplied together to retrieve the original matrix.

The intuition behind NMF is as follows:
    1. A matrix V is constructed such that:
        a. The rows of V represent one feature to be measured
        b. The columns of V represent one example containing a measurement across all features
    2. V is decomposed into W and H such that:
        a. V = WH
        b. All elements in W and H > 0
        c. W is called the features matrix
        d. H is called the coefficients matrix
    3. The columns of V can thus be represented as a linear combination of the columns in W
        a. Each column in W can thus be considers a "reduced dimension"

## 3.1 Setup
sklearn provides an implementation of NMF. This solver has the following parameters:
   1. **n_components** : k in my pseudocode. The number of components to use for the factorization
   2. **init**: the methodology of initialization. I will be using nndsvda, which seeds the optimization with a truncated singular value decomposition
   3. **solver**: the method of solving the problem. I will be using 'mu' for the multiplicitive update solver, as described in my pseduocode
   4. **beta loss**: the type of loss to be minimized. when L2 loss is minimized, the algorithm reduces to the version in my pseudocode.
   5. **tol**: the epsilon in my pseudocode. This is the convergence condition
   6. **random_state**: optional random seed for initialization
   7. **max_iter**: max number of iterations before early stop, 200 by default
   8. **alpha** optional regularization term. 0 by default
   9. **l1_ratio** optional parameter for mexing l1 and l2 norm. 0 by default. I plan to experiment with this one a bit
   10. **verbose**: prints out debug information. False by default

I will be comparing NMF to PCA in these scenerios, so I import it as well

```python
import numpy as np
from matplotlib import pyplot as plt

from sklearn.decomposition import NMF
from sklearn.decomposition import PCA
```

## 3.2 Synthetic Example of Strong Performance
For an example with strong performance, I will generate a dataset with 20 dimensions and 2 class labels. 5 of these dimensions will have a strong correlation to the class label, and the remaining 15 will be highly variant gaussian 
noise


```python
def generate_sample(label):
    noise = np.random.normal (loc = 150, scale = 25, size = 15)
    if label:
        sample = np.hstack([noise,np.random.normal (loc = 215, scale = 10, size = 5)])
        return sample
    else:
        sample = np.hstack([noise,np.random.normal (loc = 150, scale = 25, size = 5)])
        return sample

def do_test():
    #generate data
    negative = [generate_sample(0) for i in range(300)]
    positive = [generate_sample(1) for i in range(100)]
    X = negative + positive
   
    #reduce using nmf
    NMF_model = NMF(n_components=3, init='nndsvdar', solver='mu', max_iter=1000)
    NMF_model = NMF_model.fit(X)
    negN = NMF_model.transform(negative)
    posN = NMF_model.transform(positive)
    H = NMF_model.components_
   
    plt.figure()
    plt.title('NMF Scatter By Principle Component')
    plt.scatter([elem[0] for elem in negN], [elem[1] for elem in negN], c='r')
    plt.scatter([elem[0] for elem in posN], [elem[1] for elem in posN], c='g')
    plt.xlabel('Principle Component 1')
    plt.ylabel('Principle Component 2')
    plt.show()
    
    
    img = []
    posIndex = 0
    negIndex = 0
    for i in range(20):
        for j in range(20):
            if i>5 and i<10 and j>5 and j<10:
                img.append(posN[posIndex])
                posIndex+=1
            else:
                img.append(negN[negIndex])
    img = np.array(img).reshape(20, 20, 3)
    print(img.shape)
   
    for i in range(3):
        plt.figure()
        plt.title('NMF Image Visualization: PC ' + str(i))
        plt.imshow(img[:, :, i], cmap='gray')
        plt.show()
    
    #reduce using PCA
    PCA_model = PCA(n_components=3)
    PCA_model = PCA_model.fit(X)
    negP = PCA_model.transform(negative)
    posP = PCA_model.transform(positive)
    H = PCA_model.components_
   

    
    plt.figure()
    plt.title('PCA Scatter By Principle Component')
    plt.scatter([elem[0] for elem in negP], [elem[1] for elem in negP], c='r')
    plt.scatter([elem[0] for elem in posP], [elem[1] for elem in posP], c='g')
    plt.xlabel('Principle Component 1')
    plt.ylabel('Principle Component 2')
    plt.show()
    
    img = []
    posIndex = 0
    negIndex = 0
    for i in range(20):
        for j in range(20):
            if i>5 and i<10 and j>5 and j<10:
                img.append(posP[posIndex])
                posIndex+=1
            else:
                img.append(negP[negIndex])
    img = np.array(img).reshape(20, 20, 3)
    print(img.shape)
    
    for i in range(3):
        plt.figure()
        plt.title('PCA Image Visualization: PC ' + str(i))
        plt.imshow(img[:, :, i], cmap='gray')
        plt.show()
    
    
```

```python
do_test()
```

```python
myResource = NeuroDataResource(host = 'api.boss.neurodata.io',
                               token = token,
                               collection = 'collman',
                               experiment = 'collman15v2',
                               chanList = [{'name': 'annotation', 'dtype': 'uint64'},
                                           {'name': 'DAPI1st', 'dtype':'uint8'},
                                           {'name': 'DAPI2nd', 'dtype':'uint8'},
                                           {'name': 'DAPI3rd', 'dtype':'uint8'},
                                           {'name': 'EM25K', 'dtype':'uint8'},
                                           {'name': 'GABA488', 'dtype':'uint8'},
                                           {'name': 'GAD647', 'dtype':'uint8'},
                                           {'name': 'gephyrin594', 'dtype':'uint8'},
                                           {'name': 'GS594', 'dtype':'uint8'},
                                           {'name': 'MBP488', 'dtype':'uint8'},
                                           {'name': 'NR1594', 'dtype':'uint8'},
                                           {'name': 'PSD95_488', 'dtype':'uint8'},
                                           {'name': 'Synapsin647', 'dtype':'uint8'},
                                           {'name': 'VGluT1_647', 'dtype':'uint8'}
                                          ])
```

```python
cutouts = {}
for key in myResource._chanList.keys():
    if not 'EM' in key:
        print(key)
        cutout = myResource.get_cutout(chan=key, zRange=[5, 15], yRange=[1000, 2000], xRange=[1000, 2500])
        cutouts[key] = cutout
```

```python
ledger = []
chans = []
for key, value in cutouts.items():
    if not 'annotation' in key:
        ledger.append(key)
        chans.append(value)
    
volume = np.stack(chans)
```

```python
volume = np.moveaxis(volume, 0, -1).reshape([-1, 12])
volume.shape
```

```python
NMF_model = NMF(n_components=2, init='nndsvdar', solver='mu', max_iter=10000)
NMF_model = NMF_model.fit(volume)
```

```python
out = NMF_model.transform(volume)
```
