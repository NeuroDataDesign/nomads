```python
def z_transform(img):
    sigma = np.std(img)
    mu = np.average(img)
    return (img - mu)/sigma
```

```python
input_chans = {}
for key, val in cutouts.items():
    if not 'annot' in key:
        input_chans[key] = val
```

```python
def make_overlap_img(a, b):
    bin_a = np.stack([elem > threshold_otsu(elem) for elem in a])
    bin_b = np.stack([elem > threshold_otsu(elem) for elem in b])
    rgb_both = np.moveaxis(np.stack([bin_a, bin_b, np.zeros_like(bin_a)]), 0, -1)
    rgb_both = rgb_both.astype(float)
    return rgb_both
```

```python
overlap = make_overlap_img(cutouts['annotation'], cutouts['PSD95_488'])
for img in overlap:
    plt.figure()
    plt.imshow(img)
    plt.show()
```

```python
def compute_convolutional_cov(vol1, vol2, kernel_shape):
    mu_kernel = np.ones(kernel_shape)/float(np.sum(np.ones(kernel_shape)))
    e1 = convolve(vol1, mu_kernel)
    e2 = convolve(vol2, mu_kernel)
    
    e12 = convolve(np.log(np.exp(vol1)+np.exp(vol2)), mu_kernel)
    
    cov = e12 - np.log(np.exp(e1) + np.exp(e2))
    
    return cov
```

```python
conv_overlap = compute_convolutional_cov(z_transform(cutouts['PSD95_488']),
                                         z_transform(cutouts['Synapsin647']),
                                         (3, 3, 3))
```

```python
for img in conv_overlap:
    plt.figure()
    plt.imshow(img)
    plt.show()
```

```python
overlap = make_overlap_img(cutouts['annotation'], conv_overlap)
for img in overlap:
    plt.figure()
    plt.imshow(img)
    plt.show()
```

```python
def summarize(pred, gt):
    overlap = compute_overlap_array(pred, gt)
    
    plt.figure()
    plt.title('Prediction Per Gt')

    ax = plt.subplot(111)
    arr = np.array(overlap['predictionPerGt'])
       
    retrieved = np.sum([arr >= 1])
    relevant= np.sum(np.ones_like(arr))
    rec = retrieved/relevant
    
    freq = [np.sum([arr == 0]),
            np.sum([arr == 1]),
            np.sum([arr >= 2])]

    plt.bar([0, 1, 2], freq)
    ax.set_xticks([0, 1, 2])
    plt.show()
    
    plt.figure()
    plt.title('Gt Per Prediction')

    ax = plt.subplot(111)
    arr = np.array(overlap['gtPerPrediction'])
    
    tp = np.sum([arr >= 1])
    fp = np.sum([arr == 0])
    prec = tp/(fp + tp)
 
    
    freq = [np.sum([arr == 0]),
            np.sum([arr == 1]),
            np.sum([arr >= 2])]

    plt.bar([0, 1, 2], freq)
    ax.set_xticks([0, 1, 2])
    plt.show()
    return prec, rec
```

```python
bin_conv_overlap = conv_overlap > threshold_otsu(conv_overlap)
label_conv_overlap = label(bin_conv_overlap)
```

```python
prec, rec = summarize(bin_conv_overlap,
                      np.stack(cutouts['annotation']) > 0)
```

```python
print('Precision: ', prec)
print('Recall: ', rec)
```

```python
bin_annot = np.stack(cutouts['annotation']) > 0
pos = []
neg = []
for idx in np.unique(label_conv_overlap):
    if not idx == 0:
        idx_list = label_conv_overlap == idx
        if any(bin_annot[idx_list]):
            pos.append(np.sum(label_conv_overlap == idx))
        else:
            neg.append(np.sum(label_conv_overlap == idx))

print(len(pos))
print(len(neg))
```

```python
plt.figure()
plt.hist(pos, alpha=.5, label='True Positive')
plt.hist(neg, alpha=.5, label='False Positive')
plt.show()
```

```python
plt.figure()
plt.hist(pos, alpha=.5, label='True Positive', bins = range(0, 100))
plt.hist(neg, alpha=.5, label='False Positive', bins = range(0, 100))
plt.show()
```

```python
def remove_low_volume_predictions(label_img, thresh):
    keep_list = []
    for idx in np.unique(label_img):
        if not idx == 0:
            if(np.sum(label_img == idx)) > thresh:
                keep_list.append(idx)
                
    return np.isin(label_img, keep_list)
```

```python
prec_list = []
rec_list = []
f1_list = []

for i in range(1, 40):
    filtered_pred = remove_low_volume_predictions(label_conv_overlap, i)
    
    prec, rec = summarize(filtered_pred, bin_annot)
    
    prec_list.append(prec)
    rec_list.append(rec)
    f1_list.append(2 * prec * rec / (prec + rec))
    
```
