```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True) 

import numpy as np
```

```python
N = 1000
random_x = np.random.randn(N)
random_y = np.random.randn(N)
```

```python
trace = go.Scatter(
    x = random_x,
    y = random_y, 
    mode = 'markers',
    marker = dict(
    size = '9',
    color = np.random.randn(N),
    colorscale = 'Viridis',
    showscale = True
    )
)
    
```

```python
data = [trace]
```

```python
iplot(data)
```

### Errors With this Scatter Plot

The scatter plot above plots x and y points separately as opposed to plotting them as coordinates. If we had (x,y) coordinate pairs, then we should only have a 2D color scheme. **Learn how to fix this issue!**

```python
random_z = np.random.randn(N)
```

```python
trace = go.Scatter3d(
    x = random_x,
    y = random_y,
    z = random_z,
    mode = 'markers',
    marker = dict(
    size = '9',
    color = np.random.randn(N),
    colorscale = 'Viridis',
    showscale = True
    )
)
```

```python
data_1 = [trace]
```

```python
iplot(data_1)
```
