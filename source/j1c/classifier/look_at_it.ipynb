```python
with open('./api_key.txt') as f:
    token = f.readline()
host = 'api.boss.neurodata.io'
collection = 'collman'
experiment = 'M247514_Rorb_1_Site3Align2_LENS_Session1_CROP'

ndr = NeuroDataResource(host, token, collection, experiment)
```

```python
channels = [ 'GABA',
 'GAD2',
 'Gephyrin',
 'GluN1',
 'PSD95',
 'synapsin',
 'TdTomato',
 'VGlut1']
```

```python
data = np.load('./data.npy')
df = pd.read_csv('./Site3Synaptograms.csv')
```

```python
data.shape
```

```python
X = data.copy()
```

```python
X = X.swapaxes(0, 1)
```

```python
X.shape
```

```python
X = X.reshape((8, -1))
```

```python
for idx, channel in enumerate(X):
    X_ = channel.ravel()
    title = 'Histogram of intensities for {}'.format(channels[idx])
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.hist(X_, 255, log=True)
    ax.set_title(title)
    display(fig)
```
